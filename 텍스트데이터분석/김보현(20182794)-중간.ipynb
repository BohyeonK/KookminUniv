0이면 10대, 혹은 20대 이고 1이면 30대, 40대, 50대이거나 전체 연령이 고르게 분포한 경우

```python
def max_ages(x):
    if x == '오류':
        return '오류'
    else:
        m = max(x)
        n = []
        for i in range(len(x)):
            if x[i] == m:
                n.append(i)
        if 0 or 1 in n:
            y = 0
        elif 2 or 3 or 4 in n:
            y = 1
        else:
            y = 1
        print(n)
        return y
```

```python
for i in range(len(ages)):
    ages[i] = max_ages(ages[i])
```

```python
for i in range(len(ages2)):
    ages2[i] = max_ages(ages2[i])
```

# LDA

```python
from gensim.matutils import Sparse2Corpus
from gensim.corpora.dictionary import Dictionary
import re

corpus = Sparse2Corpus(x_train.T)
id2word = dict(enumerate(tfidf.get_feature_names()))
token_re = re.compile(tfidf.token_pattern)
docs = []
for copy in copy_train:
    doc = [tok.lower() for tok in token_re.findall(copy)]
    docs.append(doc)
dic = Dictionary(docs)
dic.filter_extremes(no_below=10,no_above=0.2)
bows = []
for doc in docs:
    bow = dic.doc2bow(doc)
    bows.append(bow)
```

```python
from gensim.models.ldamodel import LdaModel

model = LdaModel(corpus=corpus,id2word=id2word,num_topics=10,passes=3,iterations=100,random_state=123)
```

```python
model.show_topic(0)
```

```python
cps_test = Sparse2Corpus(x_test.T)
test_corpus = Sparse2Corpus(x_test.T)
log_pp = model.log_perplexity(test_corpus)
2**-log_pp
```

```python
dic = Dictionary()
dic.token2id = {t:i for i, t in enumerate(tfidf.get_feature_names())}
```

```python
from gensim.models import CoherenceModel

words = set(tfidf.get_feature_names())
texts = []
for j in copy_test:
    text = []
    for word in token_re.findall(j):
        word = word.lower()
        if word in words:
            text.append(word)
    texts.append(text)

coh = CoherenceModel(model=model,texts=texts,dictionary=dic,coherence='c_v')
coh.get_coherence()
```

## hyperopt

```python
from hyperopt import hp
from hyperopt.pyll.stochastic import sample
from hyperopt import STATUS_OK
from hyperopt import fmin, rand, tpe, Trials
import uuid
```

```python
space = hp.choice('model',[{'layer':'multi','lr':hp.loguniform('lr2',-7,-2),'units':hp.quniform('units',16,256,16),
                            'activation':hp.choice('activation',['relu','tanh'])}])
```

```python
def objective(h):
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(units=h['units'],activation=h['activation']))
    model.add(tf.keras.layers.Dense(units=h['units'],activation=h['activation']))
    model.add(tf.keras.layers.Dense(units=h['units'],activation=h['activation']))
    model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
    model.compile(optimizer=tf.keras.optimizers.Adam(lr=h['lr']),loss='binary_crossentropy',metrics=['accuracy'])
    model.fit(x_train.toarray(),y_train.values,epochs=3,verbose=0)
    loss, acc = model.evaluate(x_test.toarray(),y_test.values,verbose=0)
    filename = str(uuid.uuid4())
    model.save(filename)
    return {'loss':-acc,'status':STATUS_OK,'filename':filename}
```

```python
h = sample(space)
```

```python
objective(h)
```

```python
trials = Trials()
```

```python
fmin(objective,space=space,algo=tpe.suggest,max_evals=10,trials=trials)
```

```python
trials.best_trial
```

```python
trials.best_trial['result']
```

```python
best_model = tf.keras.models.load_model(trials.best_trial['result']['filename'])
```

```python
best_model.evaluate(x_test,y_test)
```
